```python
class Tree:
    def __init__(self,name, designation):
        self.name = name
        self.designation = designation
        self.children = []
        self.parent = None
        
    def add_child(self,child):
        child.parent = self
        self.children.append(child)
        
    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level
        
    def print_tree(self,property):
        if property == 'both':
            val = self.name
        if property == 'name':
            val = self.name
        else:
            val = self.designation
            
        spaces = ' ' * self.get_level()*3
        prefix = spaces+'|__'
        print(prefix+val)
        for child in self.children:
            if self.children:
                child.print_tree(property)
                
def office_tree():
    root = Tree("Nilpul","CEO")
    
    node1 = Tree('chinmai',"CTO")
    
    
    
    node2 = Tree('gels',"sdjf")
    node2.add_child(Tree('peter',"fnk"))
    node2.add_child(Tree('waqas',"kjfe"))
    
    node3 = Tree('viswa',"jkf")
    node3.add_child(Tree('dhaval',"nfb"))
    node3.add_child(Tree('abhijit',"ywge"))
    
    node4 = Tree('amir',"djfhfdwi")
    
    root.add_child(node1)
    root.add_child(node2)
    node1.add_child(node3)
    node1.add_child(node4)
    
    return root




if __name__ == '__main__':
    root = office_tree()
    root.print_tree('name')
    root.print_tree('designation')
    root.print_tree('both')
    
   
    
    
        
```
